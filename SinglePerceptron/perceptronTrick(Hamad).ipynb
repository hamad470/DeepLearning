```python
from sklearn.datasets import make_classification
X,y = make_classification(n_samples=100,n_classes=2,n_clusters_per_class=1,n_informative=1,n_redundant=0,n_features=2,hypercube=False,random_state=41,class_sep=10)
```

```python
import matplotlib.pyplot as plt 
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap ="winter",s=80)
```

```python
##imports
import numpy as np
import random
```

```python
class Perceptron:
    def __init__(self):
        self.lr =0.1
        self._coef = 0
        self._intercept = 0
    def initialize_weights(self,n_features):
        weights = np.ones(n_features)
        return weights
    def step(self,x):
        return 1 if x>=0 else 0 
        

    def fit(self,X,y):
        X = np.insert(X,0,1,axis=1)
        weights = self.initialize_weights(X.shape[1])
        for i in range(1000):
            j = random.randint(0,99)
            y_pred = self.step(np.dot(weights,X[j]))
            weights = weights + self.lr*(y[j]-y_pred)*X[j]
        self._intercept = weights[0] 
        self._coef = weights[1:]
        
        



        

obj = Perceptron()
obj.fit(X,y)
```

```python
obj._intercept
```

```python
obj._coef
```

```python
m = -obj._coef[0]/obj._coef[1]
b = -obj._intercept/obj._coef[1]
x_input = np.linspace(-3,3,100)
y_input = m*x_input + b
```

```python
plt.figure(figsize=(14,8))
plt.plot(x_input,y_input,color ="red",linewidth =3)
plt.scatter(X[:,0],X[:,1],c=y,cmap="winter")
plt.ylim(-3,2)
```
