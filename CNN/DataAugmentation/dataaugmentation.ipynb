```python
import tensorflow
from tensorflow import keras
from keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
img = image.load_img('train/cat.jpeg',target_size=(128,128))
img
```

```python
print(type(img))
```

```python
img = image.img_to_array(img)
print(type(img))
```

```python
img_batch = img.reshape(1,128,128,3)
```

```python
datagen = ImageDataGenerator(
rotation_range =30,
shear_range  =0.2,
zoom_range = 0.2,
horizontal_flip = True,
width_shift_range = 0.2,
height_shift_range  =0.2,
fill_mode = "nearest"
)
```

```python
i=0
for output in datagen.flow(img_batch,batch_size=1,save_to_dir="aug"):
    i+=1
    if i ==10:
        break
```
